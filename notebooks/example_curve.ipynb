# Curve Example
This notebook will show how to use the `fdasrsf` package to align and statistically analyze a set of curves using the SRVF framework

## Load Packages
We will load the required packages and the example data set (MPEG7)

```python
import fdasrsf as fs
import matplotlib.pyplot as plt
import numpy as np
data = np.load('../bin/MPEG7.npz',allow_pickle=True)
Xdata = data['Xdata']
```

Now we will construct a 3-D array of a set of 2-D curves of the bone shape from the MPEG7 database

```python
curve = Xdata[0,1]
n,M = curve.shape
K = Xdata.shape[1]

beta = np.zeros((n,M,K))
for i in range(0,K):
    beta[:,:,i] = Xdata[0,i]
```

## Analyze
We now will construct a `fdacurve` object

```python
obj = fs.fdacurve(beta,mode='C',N=M)
```

Next, find the Karcher mean and align the curves to the mean

```python
obj.karcher_mean()
obj.srvf_align()
```

We will now plot the results

```python
obj.plot()
```

## Shape PCA
Next construct the Karcher Covariance and compute PCA

```python
obj.karcher_cov()
obj.shape_pca()
```

Plot results

```python
obj.plot_pca()
```
